```python
stu['parental level of education'].unique()
```

```python
deg = ['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"]

sns.countplot(data=stu, x = 'parental level of education', order = deg)
plt.xticks(rotation = 90);
```

```python
deg = ['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"]
# reading score on the y axis :
sns.barplot(data=stu, x = 'parental level of education', y = 'reading score', order = deg, ci=None) # "ci" stands for "confidence interval" 
plt.xticks(rotation = 90);
```
