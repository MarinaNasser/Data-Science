## 1. Importing data and needed libiraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Marina\Downloads\fuel_econ.csv")
df.head()
```

## 2. try to make a plot like given one

4.jpg

```python
plt.figure(figsize=[10,5])

for i in df['cylinders'].unique():
    temp= df[df['cylinders']==i]
    plt.scatter(temp['co2'],temp['city'],label= i, alpha= 0.2)


plt.legend();
```

```python
plt.figure(figsize=[10,5])
sns.scatterplot(data=df, x= 'co2',y= 'city',hue = 'cylinders',alpha = 0.2);
```

## 3. try to make a plot like the given one

1.jpg

```python
plt.figure(figsize=[15,10])

sns.pointplot(data=df, x="year", y="co2", hue='fuelType', ci = None)
plt.legend(loc='lower right')
plt.title("CO2 emmission over years ")
plt.ylabel("Co2 Emmission");
```
