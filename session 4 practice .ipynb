## 3. try to make a plot like given one
- you need to choose the top 5 car brands only from data
- use `make` in x axis and `ghgScore` in y axis 

2.jpg

```python
new=df['make'].value_counts().index[:5]
```

```python
newdf=df[df['make'].isin(new)]
```

```python
newdf=newdf[newdf['VClass']!='Minicompact Cars']
```

```python
plt.figure(figsize=[15,10])
sns.countplot(data=newdf, x ='make', hue='VClass');
```

```python
plt.figure(figsize=[15,10])
sns.barplot(x=newdf["make"] , y=newdf["ghgScore"] , hue=newdf["VClass"], ci=False);
```
