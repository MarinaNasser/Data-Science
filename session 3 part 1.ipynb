### Converting numeric data to levels, but why? 


```python
dfAll['pH'].describe()
```

```python
# Bin edges that will be used to "cut" the data into groups
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01 ]
```

```python
# Labels for the four acidity level groups
bin_names = ['high' , 'moderately high' ,'medium' ,'low' ] 
```

```python
dfAll['acidity_levels'] = pd.cut(dfAll['pH'], bin_edges, labels=bin_names)

dfAll.head()
```

```python
dfAll.groupby('acidity_levels').mean()
```
