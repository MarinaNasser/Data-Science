```python
df['math score'].fillna(df['math score'].mean(), inplace=True)
```

```python
df['reading score'] = df['reading score'].replace('-', np.nan)
df['reading score'] = df['reading score'].replace('..', np.nan)
df['reading score'] = df['reading score'].astype(float)
df['reading score'].fillna(df['reading score'].mean(), inplace=True)
```

```python
df['writing score'] = df['writing score'].str.replace("%","")
df['writing score'] = df['writing score'].astype(float)
```

### 5. **EXIT TICKET ;)** 
- Create `total` column (sum of scores)
- Create  `grade` column (F, D, C, B, A)

```python
df['total'] = (df['reading score'] + df['writing score'] + df['math score'])/3
```

```python
binEdges= [0,60,70,80,90,100]
binNames= ['F','D','C','B','A']
```

```python
df['grade']= pd.cut(df['total'], binEdges, labels=binNames)
```
